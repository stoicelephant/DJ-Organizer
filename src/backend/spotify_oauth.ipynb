```python
import spotipy
from spotipy.oauth2 import SpotifyOAuth
```

```python
import os
from dotenv import load_dotenv
load_dotenv()
SPOTIPY_CLIENT_ID = os.getenv('SPOTIPY_CLIENT_ID')
SPOTIPY_CLIENT_SECRET = os.getenv('SPOTIPY_CLIENT_SECRET')
SPOTIPY_REDIRECT_URI = os.getenv('SPOTIPY_REDIRECT_URI')
```

```python
scope = "playlist-read-private"

sp = spotipy.Spotify(auth_manager=SpotifyOAuth(scope=scope))

# auth_manager = SpotifyClientCredentials()
# sp = spotipy.Spotify(auth_manager=auth_manager)

playlists = sp.current_user_playlists()
while playlists:
    for i, playlist in enumerate(playlists['items']):
        print(f"{i + 1 + playlists['offset']:4d} {playlist['uri']} {playlist['name']}")
    if playlists['next']:
        playlists = sp.next(playlists)
    else:
        playlists = None
#so we can get list of playlists and ID like this 
```

```python
#can get a list of song ids from the playlist
playlists_info = sp.playlist_items('2DePO7r7crNhiqNsUhniyV')

playlist_ids = []

for i in playlists_info['items']:
    playlist_ids.append((i['track']['id']))
```

```python
playlists_info = sp.playlist_items('2DePO7r7crNhiqNsUhniyV')

track_ids = []

while playlists_info:
    for item in playlists_info["items"]:
        if item["track"]:
            track_ids.append(item["track"]["id"])

    if playlists_info["next"]:
        playlists_info = sp.next(playlists_info)
    else:
        playlists_info = None

print(track_ids)
```
