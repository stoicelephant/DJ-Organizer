```python
import h5py
```

```python
import os

folder_path = r"C:\Users\james\projects\dj\src\files\MillionSongSubset\files"

files = os.listdir(folder_path)

tree_files = []

for i in files:
    tree_files.append(f'{folder_path}\{i}')

deeper_files = []

for i in tree_files:
    for j in os.listdir(i):
        deeper_files.append(f'{i}\{j}')

even_deeper_files = []

for i in deeper_files:
    for j in os.listdir(i):
        even_deeper_files.append(f'{i}\{j}')
#even deeper files lists all of the directories in which the h5 files can be found 
```

```python
directory_dict = {}

for i,j in enumerate(even_deeper_files):
    directory_dict[j] = os.listdir(j)

directory_dict 
#we can now see that we have a dictionary containing all of the h5 files, indexed by the enumeration of directory paths 
```

```python
file_base = list(directory_dict.keys())[0]
file_end =  directory_dict[file_base][0]
example_file = f'{file_base}\{file_end}'
```

```python
with h5py.File(example_file,"r") as f:
    print(list(f['analysis'].keys()))
```
